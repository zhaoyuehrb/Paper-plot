```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
comb = pd.read_csv('collate_genes_withER.csv')
comb
```

```python
comb = comb.sort_values(by='ER_L24_t1_logFC').reset_index(drop = True)
comb['bin'] =  comb.index // (len(comb)/5)
comb
```

```python
ax = sns.boxplot(data=comb,x='bin',y='seqlogo_logL_S15_t0')
ax.set_title('enriched at right')
```

```python
ERs = ['ER_L24_t1_logFC','ER_S15_t0_logFC','ER_S15_t1_logFC','ER_S15_t2_logFC']
logos = ['seqlogo_logL_L24_t1','seqlogo_logL_S15_t1','seqlogo_logL_S15_t0']
for ER in ERs:
    for logo in logos:
        comb = comb.sort_values(by=ER).reset_index(drop = True)
        comb['bin'] =  comb.index // (len(comb)/5)
        fig = plt.figure(figsize=(9,9))
        ax = sns.boxplot(data=comb,x='bin',y=logo)
        ax.set_title('Similarity with enriched logo \''+logo[-6:]+'\' for '+ ER + ' (enriched on the right)')
        fig.savefig('likelihood_ER/'+'Similarity with enriched logo '+logo[-6:]+' for '+ ER+'.png', dpi=100)
```
