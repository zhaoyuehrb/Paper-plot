```python
TE_files = glob.glob('../../RBC/Deseq2 Analysis/shift15/coding/csv_raw/RNA*.csv')
TE_files
```

```python
TE_files[2].split('\\')[-1][:-4]
```

```python
for file in TE_files:
    TE = pd.read_csv(file)
    keyword = file.split('\\')[-1][:-4]
    geneGroupTitle = list(geneGroup.keys())
    TISs = []
    sizes = []
    TEs = []
    for gg in geneGroupTitle:
        TEs.append(np.mean(TE.loc[TE['Unnamed: 0'].isin(geneGroup[gg]),'log2FoldChange']))
        TISs.append(np.mean(TIS.loc[TIS['AccNum'].isin(geneGroup[gg]),'Efficiency']))
        sizes.append(len(geneGroup[gg]))
    comb_df = pd.DataFrame({'GeneGroup':geneGroupTitle,'TIS':TISs,'RNA':TEs,'size':sizes})
    comb_df.to_csv(keyword+'.csv', encoding='utf-8', index=False)
```

```python
path = '../../RBC/RPKM with refseq/shift15/coding/'
for i in range(18,24):
    TE = pd.read_csv(path+'shift15_geneTXCD_RPKMoutput_'+fid['FileName'][i]+'_shift15.txt',sep='\t',na_values=['-'])
    keyword = 'RNA_t'+str(fid['Timepoint'][i])+'_set'+str(fid['Sets'][i])
    geneGroupTitle = list(geneGroup.keys())
    TISs = []
    sizes = []
    TEs = []
    for gg in geneGroupTitle:
        TEs.append(np.mean(TE.loc[TE['AccNum'].isin(geneGroup[gg]),'cdRPKM']))
        TISs.append(np.mean(TIS.loc[TIS['AccNum'].isin(geneGroup[gg]),'Efficiency']))
        sizes.append(len(geneGroup[gg]))
    comb_df = pd.DataFrame({'GeneGroup':geneGroupTitle,'TIS':TISs,'RNA':TEs,'size':sizes})
    comb_df.to_csv(keyword+'.csv', encoding='utf-8', index=False)
```
